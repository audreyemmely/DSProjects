```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('Summer-Olympic-medals.csv', encoding = 'latin-1')
```

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
db = data['Country'].value_counts().sort_values(ascending = False)
top15 = db[:15]
top15.plot(kind = 'bar', figsize = (10, 4), color = 'r')
plt.xticks(rotation = 70)
plt.title('All time medals of top 15 countries')
```

```python
db = data['City'].value_counts().sort_values(ascending = False)
top15 = db[:15]
top15.plot(kind = 'bar', figsize = (10, 4), color = 'g')
plt.xticks(rotation = 70)
plt.title('All time medals of top 15 citties')
```

```python
db = data['Athlete'].value_counts().sort_values(ascending = False)
top15 = db[:15]
top15.plot(kind = 'bar', figsize = (10, 4), color = 'b')
plt.xticks(rotation = 70)
plt.title('All time medals of top 15 athletes')
```

```python
data['Medal'].value_counts()
```

```python
sns.countplot(data.Medal)
plt.title('Total number of medals')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.countplot(y = "Medal", hue = "Gender", data = data)
plt.title('Number of medals by gender')
plt.show()
```

```python
#USA medals by year 
usa = data[data.Country_Code=='USA']
sns.countplot(x= "Year", data = usa)
plt.title("USA total medals per year")
plt.xticks(rotation = 90)
plt.show()
```

```python
#USA medals by type
sns.countplot(x = 'Year', hue = 'Medal', data = usa, palette =('gold', 'brown', 'silver'))
plt.title('USA medals by year and medal type')
plt.show()
```

```python
data['Sport'].value_counts()
```

```python
sns.countplot(data.Sport)
plt.title('Number of medals by sports')
plt.xticks(rotation = 90)
plt.show()
```

```python
data['Gender'].value_counts()
```

```python
sns.countplot(x = "Gender", data = data, palette = 'Set2')
plt.title('Women vs Men')
plt.show()
```

```python
#Number of men vs women represented by year
sns.countplot(x = 'Year', hue = 'Gender', data = data, palette = ('red', 'blue'))
plt.title('Total medals awarded: Men vs Women')
plt.show()
```
